```python
fake = Faker()

def data1_gen(samples):
  name =[]
  address = []
  phone_number =[]
  email = []
  job_title = []
  credit_card_number = []
  uuid = []
  amount = []

  for _ in range(samples):
    name.append(fake.name())
    address.append(fake.address())
    phone_number.append(fake.phone_number())
    email.append(fake.email())
    job_title.append(fake.job())
    credit_card_number.append(fake.credit_card_number())
    uuid.append(fake.uuid4())
    amount.append(fake.pydecimal(left_digits=4, right_digits=2, positive=True))

  return name, address, phone_number, email, job_title, credit_card_number, uuid, amount
```

```python
name, address, phone_number, email, job_title, credit_card_number, uuid, amount = data1_gen(10000)
data = {'name':name, "address": address, "phone_number": phone_number, "email":email, "job_title": job_title, "credit_card_number":credit_card_number, "uuid":uuid,
        "amount":amount}
df_data = pd.DataFrame(data)
```

```python
import random

def generate_phone_number(start_number, len_of_phone_num, tot_req_values):
    number = start_number
    store = []

    def generate_number(len_of_phone_num, number):
      num = number
      for _ in range(len_of_phone_num):  # Generate the remaining 9 digits
          num += str(random.randint(0, 9))
      return num

    # Generate 5 random phone numbers
    for _ in range(tot_req_values):
          store.append(generate_number(len_of_phone_num, number))

    return store
```
